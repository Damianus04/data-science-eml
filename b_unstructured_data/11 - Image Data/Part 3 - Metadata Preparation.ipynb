# How to get all filenames?

```python
from glob import glob
```

```python
fname = glob("data/*/*.png")
fname[:10]
```

# X = filename --> read --> flattened --> concat/gabung

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = [plt.imread(f).flatten() for f in fname]
X = np.array(X)
```

```python
X.shape
# (row, column)
```
