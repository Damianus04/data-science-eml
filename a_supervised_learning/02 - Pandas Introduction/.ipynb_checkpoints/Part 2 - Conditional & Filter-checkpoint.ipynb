# Map, Apply & Transform

#### Map ****

```python
# Mapping shelf dari angka menjadi kategori atas bawah tengah
df_cereal.shelf = df_cereal.shelf.map({1: "bawah",
                                      2: "tengah", 
                                      3: "atas"})
df_cereal
```

```python
# Mengkonversi satuan fat ke dalam satuan mg
def convert_to_mg(x):
    return x * 1000
```

#### Apply ***

```python
df_cereal.fat = df_cereal.fat.apply(convert_to_mg)
df_cereal
```

```python
# Mengkonversi huruf di vitamins menjadi lowercase
def convert_to_lowercase(x):
    return x.lower()

#alternatif lain adalah dengan annonymous function atau lambda
# df_cereal.vitamins = df_cereal.vitamins.apply(lambda x: x.lower())
```

```python
df_cereal.vitamins = df_cereal.vitamins.apply(convert_to_lowercase)
```

```python
df_cereal
```

#### Transform

```python
# melakukan normalisasi
df_cereal.rating.transform(lambda x: x/x.sum())
df_cereal
```

# One-hot Encoding / Dummy Variable ***
handling categorical label

1. Ordinal Encoding & Label Encoding
    - menomori setiap data category
    - kelemahannya adalah tidak fair karena ada label dinomori 0 atau 1 atau 2
    - efektif dipakai jika label nya menandakan urutan [binary (gender), ranking description (tidak bagus, bagus, sangat bagus)]
    

2. One Hot Encoding
    - membuat column baru
    - setiap data yang ada diberi angka 1, jika tidak ada maka 0

```python
df_cereal = pd.get_dummies(df_cereal, columns=['vitamins'])
df_cereal.head()
```

# Binning ***

```python
df_cereal["grade"] = pd.cut(df_cereal.rating, bins=[0, 40, 70, 100], labels=["jelek", "lumayan", "enak"])
df_cereal.head()
```
