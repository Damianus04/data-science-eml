##### Style
1. Matlab Style
2. Matplotlib Style

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100) #angka ketiga menentukan jumlah titik untuk graphic nya
y = np.sin(X)
ycos = np.cos(X)
```

# 1. MATLAB Style

```python
plt.plot(X, y)
```

# 2. MATPLOTLIB Style

```python
# figure
fig = plt.figure()

# axis
ax = plt.axes()

# curva
ax.plot(X, y)
```

# Developing Standard Plot

## MATLAB Style 

```python
#figure
plt.figure(figsize=(8, 6))

#box
plt.plot(X, y, "y--", linewidth=4, label="sinus")
plt.plot(X, ycos, "r--", linewidth=4, label="cosinus")


#angka untuk axis x dan y
plt.xlim(0,15)
plt.ylim(-2, 2)

#title
plt.title("Trigonometri")
plt.xlabel("x")
plt.ylabel("y")
plt.legend() #label harus dibuat dulu di part box 
```

# MATPLOTLIB Style

```python
#figure
fig = plt.figure(figsize=(8, 6))

# axis
ax = plt.axes()

#box
ax.plot(X, y, "c+", linewidth=4, label="sinus")
ax.plot(X, ycos, "r--", linewidth=4, label="cosinus")

#angka untuk axis x dan y
ax.set_xlim(0,15)
ax.set_ylim(-2, 2)

#title
ax.set_title("Trigonometri")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend() #label harus dibuat dulu di part box 

# penggunaan set bisa seperti ini
# ax.set(xlim=(0,15), ylim=(-2,2), xlabel="x", ylabel="y", title="2 fungsi trig")
```

    'b' = warna biru, solid line
    'ro' = warna merah, marker bulat
    'g-' = warna hijau, solid line
    '--' = warna default, garis dashed
    'k^:' = warna hitam, marker segitiga, garis dotted
    'c+-' = warna cyan, marker +, solid line

# Subplot

## MATLAB Style

```python
plt.subplot(221)
plt.plot(X, y)

plt.subplot(222)
plt.plot(X, ycos, 'r')

plt.subplot(223)
plt.plot(X, ycos, 'r')

plt.subplot(2, 2, 4)
plt.plot(X, y)
#box
plt.plot(X, y, "y--", linewidth=4, label="sinus")
plt.plot(X, ycos, "r--", linewidth=4, label="cosinus")
#angka untuk axis x dan y
plt.xlim(0,15)
plt.ylim(-2, 2)
#title
plt.title("Trigonometri")
plt.xlabel("x")
plt.ylabel("y")
plt.legend() #label harus dibuat dulu di part box 
```

## MATPLOTLIB Style

```python
# Tentukan ukurannya figure
fig, axes = plt.subplots(2, 2, figsize = (8, 8))

# Tentukan axis nya
# Cara mengakses subplot adalah dengan method array
axes[0, 1].plot(X, y)
axes[1, 1].plot(X, y)
```

# Scatter Plot

```python
#jika delimiter ";", set dengan delimiter=";"

df_circle = pd.read_csv("data/circle_data.csv", delimiter=";")
df_circle.head()
```

```python
plt.scatter(df_circle.x, df_circle.y)
```

```python
#karena resolusi tidak sama, maka terlihat oval
# solusinya
plt.scatter(df_circle.x, df_circle.y, c="r", s=50)
plt.axis("equal");
# ";" diletakkan di akhir untuk menghilangkan print(hasil angka plot)
```

#### Bubble Plot

```python
plt.scatter(df_circle.x, df_circle.y, c=df_circle.radius, s=df_circle.radius*50, cmap="hsv")
#c=color, s=size, cmap=color map
plt.colorbar() #display gradasi warna
plt.axis("equal");
```
