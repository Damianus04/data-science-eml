```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_linear = pd.read_csv("data/linear.csv")
df_linear.head()
```

```python
# Scatter Plot

# Title
plt.title('linear plot')

df_linear1 = df_linear[df_linear.tipe == 'data1']
df_linear2 = df_linear[df_linear.tipe == 'data2']

plt.scatter(df_linear1.X, df_linear1.y, label='data1')
plt.scatter(df_linear2.X, df_linear2.y, label='data2')

#axis label
plt.xlabel('x')
plt.ylabel('y')

#legend
plt.legend();
```

# Statistical Summary

```python
df_cereal = pd.read_csv("data/cereal.csv", index_col="name")
df_cereal.head()
```

##### Pairplot

```python
sns.pairplot(df_cereal, hue="shelf", vars=["calories", "fat", "rating"]);
```

# Persebaran Data untuk Categorical Data


### Persebaran

```python
#catplot=categorical plot
sns.catplot("mfr", "rating", data=df_cereal) 
```

```python
#boxplot
sns.boxplot("mfr", "rating", data=df_cereal) 
```

```python
# tanda diamond adalah outlier data
# garis atas=maximum, garis bawah=minimum, garis tengah=median
```

### Frekuensi

```python
sns.countplot("mfr", data=df_cereal)
```

```python
#hue/membedakan berdasarkan rak
sns.countplot("mfr", data=df_cereal, hue='shelf')
```
